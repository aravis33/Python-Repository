# 02 Combine data

```python
# determine best way to merge data sets
df_prods.head()
```

```python
orders_products_combined.head()
```

```python
# product_id is the only column that's the same, so it should be the key
# columns not the same, so must use merge
```

```python
# remove Unnamed: 0
df_prods.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
# remove _merge
orders_products_combined.drop(columns = ['_merge'], inplace = True)
```

```python
df_prods.head()
```

```python
orders_products_combined.head()
```

```python
# merge
orders_products_merged = orders_products_combined.merge(df_prods, how = 'outer', on = 'product_id', indicator = True)
```
